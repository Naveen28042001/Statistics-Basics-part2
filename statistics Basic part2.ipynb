```python
Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?


1.Mean:
    Ratio of sum of values to the total number of values in the data set
2.medain:
    Median is the middle value on the certain range of values.
3.Mode:
    Mode measures the most frequent values(Highest frequency) in the dataset.
    
Differences:
1.Sensitivity to Outliers:
  The mean is highly sensitive to extreme values, as it considers every data point in its calculation.
  The median is less sensitive to extreme values, as it is based on the position of values in the ordered dataset.
  The mode is not influenced by extreme values at all, as it is solely based on frequency.
2.Applicability:
  The mean is appropriate for symmetric distributions without outliers.
  The median is suitable for skewed distributions or datasets with outliers.
  The mode is useful for identifying the most frequent value in any distribution.
3.Uniqueness:
  The mean and median are unique values in a dataset.
  A dataset may have no mode, one mode (unimodal), or more than one mode (multimodal).
    
Use in Describing Central Tendency:
Choosing a Measure:
 The choice between mean, median, and mode depends on the characteristics of the dataset and the nature of the distribution.
 For symmetrical distributions without outliers, the mean may be preferred.
 For skewed distributions or datasets with outliers, the median is often a better measure of central tendency.
 The mode is useful for identifying the most frequent value or values in a dataset.
```

```python
Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.

Range:
Definition: The range is the difference between the maximum and minimum values in a dataset.
Use: It gives a simple and quick measure of the overall spread but is sensitive to outliers.
Example: Consider the dataset [10, 15, 12, 18, 25]. The range is 25−10=15

Variance:
Definition: Variance measures how far each data point in the set is from the mean. It is the average of the squared differences from the mean.
Use: Provides a more comprehensive measure of dispersion, but it involves squared deviations.
Example: For the dataset [10, 15, 12, 18, 25], the mean is 16. The squared differences from the mean are 
(6^2,-1^2,-4^2,2^2,9^2). The variance is the average of these squared differences.

Standard Deviation:
Definition: The standard deviation is the square root of the variance. It is often preferred because it is in the same units as the original data.
Use: Provides a more interpretable measure of dispersion than variance.
Example: Using the same dataset, the standard deviation is the square root of the variance.
```

```python
Q8. What do you understand about skewness in data?

Skewness:
    Skewness is the kind of statistical measure inthe data science whis describes about the asymmetry of the distribution.
    It aslo provides insights about the shape of the distribution.
    It gives an measure or can quantify to extent distribution differ from normal distribution.
    
Types of Skewness:
1.Zero Skewness:
    The distribution is perfectly symmetrical.
2.Negative/Left Skewness:
    The tail on the left hand side is much longer than the tail on the right hand side.
    Most of the data lies on the right hand side of the distribution.
       "Mean<Median<Mode"
    Example:
        Most of the students in the school got more marks while less students got low marks.
3.Positive/Right Skewness:
    The tail on the right hand side is much longer than the tail on the left hand side.
    Most of the data lies on the left hand side of the distribution.
       "Mean>Median>Mode"
    Example:
        Most of the employees in the company got low salary while few of the employees got high salarty.
    
```

```python
Q10. Explain the difference between covariance and correlation. How are these measures used in
statistical analysis?

Covariance:
   Covariance is a measure of how much two variables change together. 
   It indicates whether an increase in one variable corresponds to an increase or decrease in another variable.
   It gives idea about the directional relationship.
   It does not give idea about strength.
   It has unit.
   values of covariance - (-∞ to +∞)
Correlation:
    Correlation is a standardized measure of the strength and direction of the linear relationship between two variables. 
    It normalizes covariance, providing a scale-independent measure.
    It is unitless.
    Values of Correlation - (-1 to +1)
Use in Statistical Analysis:
Covariance:
  Covariance is used to understand the direction of the relationship between two variables.
  However, its magnitude is not easily interpretable, and it depends on the scales of the variables.
Correlation:
  Correlation is widely used in statistical analysis because it provides a standardized measure of the strength and direction of a linear relationship.
  It allows for comparisons between different pairs of variables, as it is not affected by the scales of the variables.
  Correlation is valuable in fields like finance, economics, biology, and social sciences for studying relationships between variables.
```

```python
Q12. For a normal distribution data what is the relationship between its measure of central tendency?

For a normal distribution, there is a specific relationship between its measures of central tendency: mean, median, and mode. In a normal distribution, these measures are all equal and located at the center of the distribution. This is a key characteristic of normal distributions, and it contributes to their symmetry. Here are the details:

Mean (μ):
  The mean of a normal distribution is located at the center of the distribution.
  For a perfectly symmetric normal distribution, the mean is the exact center, and the distribution is symmetrically balanced around the mean.
Median:
  The median of a normal distribution is also at the center.
  In a perfectly symmetric normal distribution, the mean and median are the same, right at the center of the distribution.
Mode:
  For a normal distribution, every point on the distribution is a mode, meaning it is the most frequently occurring value.
  The mode is also at the center of the distribution in a symmetric normal distribution.
In summary, for a normal distribution:
    Mean=Median=Mode

This equality holds for a perfectly symmetric normal distribution. 
However, it's important to note that in real-world scenarios, data may not always perfectly follow a normal distribution, and there might be slight differences between the mean, median, and mode. 
But in a truly normal distribution, these measures are equal and located at the center of the distribution.
```

```python
Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

Outliers can have a significant impact on measures of central tendency and dispersion. Central tendency measures include the mean, median, and mode, while dispersion measures include the range, variance, and standard deviation. Outliers are extreme values that differ significantly from the majority of the data points in a dataset.

Effects on Measures of Central Tendency:
1.Mean:
  Outliers can heavily influence the mean because the mean is sensitive to extreme values.
  If there are high-value outliers, the mean is pulled towards the higher end; if there are low-value outliers, the mean is pulled towards the lower end.
2.Median:
  The median is less affected by outliers because it is not sensitive to extreme values.
  It represents the middle value when the data is ordered, so extreme values do not impact it as much.
3.Mode:
  The mode is generally not influenced by outliers because it represents the most frequently occurring value.
  However, in some cases, outliers might create additional modes.
Effects on Measures of Dispersion:
1.Range:
  Outliers can significantly affect the range since it is the difference between the maximum and minimum values.
  If there are outliers, especially if they are extreme, the range can be substantially larger.
2.Variance and Standard Deviation:
  Outliers can greatly influence the variance and standard deviation because these measures involve squared differences from the mean.
  Squaring the differences amplifies the impact of outliers on these measures.
  The presence of outliers increases the spread, leading to higher variance and standard deviation.
Example:
  Consider the dataset: [10, 15, 12, 18, 25, 100].

Without Outlier:

Mean = 16
Standard Deviation ≈ 5.7

With Outlier:

Mean = 28.33
Standard Deviatio ≈ 31.6
In this example, the outlier (100) has significantly impacted both the mean and standard deviation, making them higher in the presence of the outlier.





```
